```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
```

```python
for path in paths:
    df = pd.read_csv(path)
    X = df.drop(["file","animal"],axis=1)
    y = df["animal"]
    pca = PCA(n_components=len(X.columns))
    pca.fit(X)
    print("Dataset: "+path)
    print(pca.explained_variance_ratio_)
    print("")
```

Now we set a threshold on the amount of variance explained by each component, get those and perform a scatter matrix :)

```python
THR = 0.05
for path in paths:
    df = pd.read_csv(path)
    X = df.drop(["file","animal"],axis=1)
    y = df["animal"]
    pca = PCA(n_components=len(X.columns))
    pca.fit_transform(X)
    print()
    N_useful_comp = len(np.where(pca.explained_variance_ratio_ > THR)[0])
    X = np.matrix(X)
    X = pd.DataFrame(X[:,0:N_useful_comp])
    print(path)
    print("Components Used expl. var. ratio : ",pca.explained_variance_ratio_)
    print("N_useful_comp",N_useful_comp)
    ignore_out = pd.plotting.scatter_matrix(X,c=y,figsize = [8,8])
    plt.show()

    #save df to file
    X["animal"] = y #add label column back
    X.to_csv(path[:-4]+"_PCA"+".csv")
```
