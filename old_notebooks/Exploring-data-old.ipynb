# Exploring the data
___

We created four datasets with different ways of analyzing our data but we want to know how well some frequencies differenciate one from another.

To do that we will plot a heat map for cats vs dogs, higher temp will be cats, lower dogs. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
paths = ["dataframes_mean/bands_lin.csv","dataframes_mean/bands_nolin.csv","dataframes_mean/freq_raw.csv","dataframes_mean/tonal_pitches.csv"]
for path in paths:
    df = pd.read_csv(path) #read df
    df = df.drop(["file"],axis=1)
    heat_mtx = np.zeros((df.shape[1]-1,df.shape[1]-1),dtype= np.float64)
    for i,row in df.iterrows():
        if(row["animal"]==0):
            for a in range(df.shape[1]-1):
                row
                heat_mtx[a,:] += np.array(row)[1:]
                heat_mtx[:,a] += np.array(row)[1:]
        if(row["animal"]==1):
            for a in range(df.shape[1]-1):
                heat_mtx[a,:] -= np.array(row)[1:]
                heat_mtx[:,a] -= np.array(row)[1:]
    im = plt.matshow(heat_mtx, cmap=plt.cm.hot, aspect='auto')
    plt.colorbar(im)
    plt.show()
```

```python
paths = ["dataframes/bands_lin.csv","dataframes/bands_nolin.csv","dataframes/freq_raw.csv","dataframes/tonal_pitches.csv"]
for path in paths:
    df = pd.read_csv(path) #read df
    df = df.drop(["file"],axis=1)
    heat_mtx = np.zeros((df.shape[1]-1,df.shape[1]-1),dtype= np.float64)
    for i,row in df.iterrows():
        if(row["animal"]==0):
            for a in range(df.shape[1]-1):
                heat_mtx[:,a] += np.array(row)[1:]
        if(row["animal"]==1):
            for a in range(df.shape[1]-1):
                heat_mtx[a,:] -= np.array(row)[1:]
    im = plt.matshow(heat_mtx, cmap=plt.cm.hot, aspect='auto')
    plt.colorbar(im)
    plt.show()
```
